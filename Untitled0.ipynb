# EDA-Superstore-Sales

## Project Overview
This project analyzes the Kaggle Superstore Sales dataset, focusing on three areas:
product profitability, regional performance, and customer segmentation.

## Key Findings

1. **Product Analysis**: Technology generates the highest profit at $145,454 â
   7.8x more than Furniture. It is recommended to reduce investment in Furniture
   and reallocate resources toward Technology and Office Supplies.

2. **Regional Analysis**: The West region leads in both total sales ($725,457)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')
df.head()
```

```python
print(df.isnull().sum())
```

```python
category_profit = df.groupby('Category')['Profit'].sum()
print(category_profit)
```

```python
category_profit.plot(kind='bar', figsize=(8, 5), color=['#e74c3c', '#3498db', '#2ecc71'])
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit (USD)')
plt.xticks(rotation=0)

plt.show()
```

```python
region_data = df.groupby('Region')[['Sales', 'Profit']].sum()
print(region_data)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
region_data[['Sales', 'Profit']].plot(kind='bar', ax=ax1, color=['#3498db', '#e67e22'])
ax1.set_xlabel('Region')
ax1.set_ylabel('Amount (USD)')
ax1.set_xticklabels(region_data.index, rotation=0)

ax2 = ax1.twinx()
region_data['Profit_Rate'] = region_data['Profit'] / region_data['Sales'] * 100
region_data['Profit_Rate'].plot(kind='line', ax=ax2, color='red', marker='o', linewidth=2)
ax2.set_ylabel('Profit Rate (%)')

plt.title('Sales vs Profit vs Profit Rate by Region')
plt.show()
```

```python

customer_data = df.groupby('Customer Name').agg(
    Total_Sales = ('Sales', 'sum'),
    Order_Count = ('Order ID', 'nunique')
).sort_values('Total_Sales', ascending=False)

print(customer_data.head(10))
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(customer_data['Order_Count'],
            customer_data['Total_Sales'],
            alpha=0.5,
            color='#3498db')

top5 = customer_data.head(5)
for name, row in top5.iterrows():
    plt.annotate(name,
                xy=(row['Order_Count'], row['Total_Sales']),
                fontsize=8)

plt.title('Customer Value Map')
plt.xlabel('Order Count (Frequency)')
plt.ylabel('Total Sales (Value)')
plt.show()
```

Based on this analysis report, three key insights emerge from this analysis:
1. Sales Performance by Category: Technology sales were robust, reaching $145,454.95. While Office Supplies also performed impressively, Furniture sales were notably low at only $18,451.27âmeaning Technology sales are 7.8 times higher than those of Furniture. We should consider reducing investment in the furniture segment to reallocate capital toward Technology and Office Supplies. This will allow us to leverage our competitive edge, enhance market influence, build brand awareness, and eliminate unnecessary expenditures.
2. Regional Profit Margins: The profit margin in the Central region stands at 7.9%, which is significantly lower than the other three regions. This suggests a potential price war aimed at capturing market share-similar to how Tesla has competed in the European market by cutting prices.
3. Customer Segmentation: Our high-value customer base is limited; most clients fall into the mid-range for both purchase frequency and spending. I recommend implementing VIP services for customers with a total spend exceeding $10,000 to foster loyalty and a sense of belonging through regular relationship management. For other customers, we should focus on regular product updates and promotional campaigns to improve retention and repurchase rates.
